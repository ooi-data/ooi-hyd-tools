```python
import obspy as obs
import matplotlib.pyplot as plt
```

```python
def gap_analysis(url):
    st = obs.read(url)
    _ = st.plot()
    print(st[0].stats)

    gap_list = []
    for i in st.get_gaps():
        #print(i[6])
        gap_list.append(i[6])

    print(len(gap_list))
    plt.hist(gap_list, bins=100)
    
```

```python
gap_analysis("https://rawdata.oceanobservatories.org/files/CE02SHBP/LJ01D/HYDBBA106/2025/06/27/OO-HYEA2--YDH-2025-06-27T05:07:00.000000Z.mseed")
```
